```python
df0=df[((df['Full Time'] == 0) & (df['Part Time'] == 0))]
df0.head()
```

```python
temp=pd.DataFrame()
temp['SS']=df0['SS']
temp['NE/NT']=df0['NE/NT']
temp['NT']=df0['NT']
temp['NE']=df0['NE']
temp.sort_values(by=['SS'],ascending=[False],inplace=True)
temp
```

# # NE/NT and NE

```python
import numpy as np
X=np.column_stack([temp['NE/NT'],temp['NE']])
Y=np.array(temp['SS'])
Y=Y.reshape((10,1))
print(X.shape)
print(Y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X, Y)
```

```python
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
```

```python
model=[]
for i in range(len(temp)):
    test=np.array([temp.iloc[i][1],temp.iloc[i][3]])
    test=test.T
    test=test.reshape((2,1))
    m=np.dot(lr.coef_,test)
    model.append(m[0][0]+lr.intercept_)
temp['model']=model
temp['dif_NE']=temp['SS']-temp['model']
temp
```
