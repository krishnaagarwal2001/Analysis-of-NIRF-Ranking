```python
import pandas as pd
Scores=pd.read_csv('Scores_Engineering_2021.csv')
Data=pd.read_csv('Data_Engineering_2021.csv')
```

```python
Data.head()
```

```python
Scores.head()
```

```python
len(Scores)
```

```python
len(Data)
```

```python
i=0
j=0
drop=[]
while(i<len(Scores) and j<len(Data)):
    if(Scores.iloc[i][0]!=Data.iloc[j][0]):
        drop.append(j)
        i=i-1
    i=i+1
    j=j+1
Data.drop(Data.index[drop],inplace=True)
```

```python
len(Data)
```

```python
len(Scores)
```

```python
Scores.drop('Name',axis=1,inplace=True)
Scores.head()
```

```python
df=pd.concat([Data,Scores],axis=1)
df.head()
```

```python
Parameters=['SS','FQE','FRU','PU','QP','IPR','FPPP','GPH','GUE','MS','GPHD','RD','WD','ESCS','PCS','PR']
df.drop(Parameters,axis=1,inplace=True)
df.head()
```

```python
import math
calc=[]
for i in range(len(df)):
    sum=df.iloc[i][3]+df.iloc[i][5]+df.iloc[i][6]
    if(math.isnan(sum)):
        calc.append(0)
    else:
        calc.append((15*(df.iloc[i][4])/sum))
        
df['calc']=calc
```

```python
temp=pd.DataFrame()
temp['FSR']=df['FSR']
temp['calc']=df['calc']
temp.sort_values(by=['FSR'],ascending=False,inplace=True)
print(temp.to_string())
```

```python
import matplotlib.pyplot as plt
df.plot.scatter(y='calc',x='FSR')
plt.show()
```
