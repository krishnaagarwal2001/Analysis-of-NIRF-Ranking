```python
import pandas as pd
Scores=pd.read_csv('Scores_Engineering_2021.csv')
Data=pd.read_csv('FRU_Engineering_2021.csv')
df=pd.merge(Data,Scores,on="Name")
pd.set_option('display.float_format', lambda x: '%.2f' % x)
Parameters=['SS','FSR','FQE','PU','QP','IPR','FPPP','GPH','GUE','MS','GPHD','RD','WD','ESCS','PCS','PR']
df=df.drop(Parameters,axis=1)
df['Annual_Expenditure_per_Student']=(df['Annual_Expenditure_2019']+df['Annual_Expenditure_2018']+df['Annual_Expenditure_2017'])/(df['NT']+df['Full Time Phd'])
df['Operational_Expenditure_per_Student']=(df['Operational_Expenditure_2019']+df['Operational_Expenditure_2018']+df['Operational_Expenditure_2017'])/(df['NT']+df['Full Time Phd'])
df=df.iloc[:, [0,1,2,9,10,11,3,4,5,6,7,8]]
df.to_csv('FRU.csv',index=False)
df.head(10)
```
