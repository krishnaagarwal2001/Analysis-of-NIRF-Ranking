```python
import pandas as pd
Scores=pd.read_csv('Scores_Engineering_2021.csv')
Data=pd.read_csv('GPHD_Engineering_2021.csv')
df=pd.merge(Data,Scores,on="Name")
Parameters=['SS','FRU','FSR','FQE','PU','QP','IPR','FPPP','GPH','GUE','MS','RD','WD','ESCS','PCS','PR']
df=df.drop(Parameters,axis=1)
df.to_csv('GPHD.csv',index=False)
df.head(10)
```

```python
df['Total']=df['F2019']+df['F2018']+df['F2017']
df=df[df['Total']!=0]
df.to_csv('GPHD.csv',index=False)
df['ratio']=df['GPHD']/df['Total']
df.head(10)
```

```python
df['root']=df['Total']**(1/2)
df.to_csv('GPHD.csv',index=False)
df.head(10)
```

```python
import numpy as np
print(df.shape)
x_train=df.iloc[:,10:11]
y_train=df.iloc[:,7]
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_train.shape)
print(y_train.shape)
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_train, y_train)
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
print("Training set score: {:.2f}".format(lr.score(x_train, y_train)))
```
