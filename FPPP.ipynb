```python
import pandas as pd
Scores=pd.read_csv('Scores_Engineering_2021.csv')
Data=pd.read_csv('FPPP_Engineering_2021.csv')
df=pd.merge(Data,Scores,on="Name")
Parameters=['SS','FRU','FSR','FQE','PU','QP','IPR','GPHD','GPH','GUE','MS','RD','WD','ESCS','PCS','PR']
df=df.drop(Parameters,axis=1)
df.to_csv('FPPP.csv',index=False)
df.head(7)
```

```python
DF=df.drop(['R2019','R2018','R2017','C2018','C2019','C2017'],axis=1)
df.head(7)
```

```python
DF.head(7)
```

```python
DF['S']=(7.5*DF['Total Research']+25*DF['Total Consultancy'])/10000000000
```

```python
DF.head(7)
```

```python
DF.to_csv('temp.csv',index=False)
```

```python
temp=DF
temp['Total Research']=7.5*temp['Total Research']
temp['Total Consultancy']=2.5*temp['Total Consultancy']
import numpy as np
# print(df.shape)
x_train=temp.iloc[:,1:2]
y_train=temp.iloc[:,3]
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_train.shape)
print(y_train.shape)
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_train, y_train)
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
print("Training set score: {:.2f}".format(lr.score(x_train, y_train)))
```
